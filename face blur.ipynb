```python
import cv2
cap= cv2.VideoCapture(0)
model=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def face_blur(photo):
    face=model.detectMultiScale(photo)
    if len(face) != 0:
        for nface in face:
            crop_photo = cv2.blur(photo[nface[1]:nface[1]+nface[3],nface[0]:nface[0]+nface[2]],(20,20))
            photo[nface[1]:nface[1]+nface[3],nface[0]:nface[0]+nface[2]] = crop_photo
        return photo
    else:
        return photo
```

```python
while True:
    ret,photo=cap.read()
    cv2.imshow('Face',face_blur(photo))
    if cv2.waitKey(10) == 13:
        break
cv2.destroyAllWindows()
```

```python
cap.release()
```
